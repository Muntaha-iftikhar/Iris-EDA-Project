```python
# ðŸ“Œ 1. Import Required Libraries
# -------------------------------
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# For beautiful plots
sns.set(style="whitegrid")
plt.rcParams["figure.figsize"] = (8, 5)
```

```python
df = sns.load_dataset("iris")

print("Dataset Loaded Successfully!")
print("-" * 60)
```

```python
# Pairplot (Most Important)
# -------------------------------
sns.pairplot(df, hue="species", markers=["o", "s", "D"])
plt.suptitle("Pairplot of Iris Dataset", y=1.02)
plt.show()
```

```python
# ðŸ“Œ 7. Scatter Plots
# -------------------------------
plt.figure(figsize=(7,5))
sns.scatterplot(
    data=df,
    x="sepal_length",
    y="sepal_width",
    hue="species",
    style="species",
    s=80
)
plt.title("Scatter Plot: Sepal Length vs Sepal Width")
plt.show()
```

```python
plt.figure(figsize=(7,5))
sns.scatterplot(
    data=df,
    x="petal_length",
    y="petal_width",
    hue="species",
    style="species",
    s=80
)
plt.title("Scatter Plot: Petal Length vs Petal Width")
plt.show()
```

```python
# ðŸ“Œ 8. Histograms
# -------------------------------
df.hist(bins=15, figsize=(12, 8), edgecolor="black")
plt.suptitle("Histograms of All Features")
plt.show()
```

```python
# ðŸ“Œ 9. Boxplots (Outlier Checking)
# -------------------------------
plt.figure(figsize=(10,6))
sns.boxplot(data=df)
plt.title("Boxplot of Iris Features (Checking Outliers)")
plt.show()
```

```python
# ðŸ“Œ 10. Violin Plots (Advanced)
# -------------------------------
plt.figure(figsize=(10,6))
sns.violinplot(data=df)
plt.title("Violin Plot of Iris Features")
plt.show()
```

```python
# ðŸ“Œ 11. Correlation Heatmap (Numeric Only)
plt.figure(figsize=(8,6))
sns.heatmap(
    df.corr(numeric_only=True),
    annot=True,
    cmap="coolwarm",
    linewidths=0.5
)
plt.title("Correlation Heatmap of Iris Dataset")
plt.show()
```
